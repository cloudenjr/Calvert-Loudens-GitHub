# Using StringIO from the io module to Read Data

```python
data = """\
Sample Animal Intelligence
1 Dog Smart
2 Dog Smart 
3 Cat Dumb
4 Cat Dumb
5 Dog Dumb
6 Cat Smart
"""
```

```python
dframe = pd.read_table(StringIO(data),sep='\s+')
```

```python
dframe
```

```python
# Cross-Tabulation table is just a frequency table
pd.crosstab(dframe.Animal,dframe.Intelligence,margins=True)
```
