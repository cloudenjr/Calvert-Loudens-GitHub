```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
animals = DataFrame(np.arange(16).reshape(4,4),columns=['W','X','Y','Z'],index=['Dog','Cat','Bird','Mouse'])

animals
```

```python
animals.loc[1:2,['W','Y']] = np.nan                                                                             
```

```python
animals
```

```python
behavior_map = {'W':'good','X':'bad','Y':'good','Z':'bad'}
```

```python
animal_col = animals.groupby(behavior_map,axis=1)

animal_col.sum()
```

```python
behave_series = Series(behavior_map)

behave_series
```

```python
animals.groupby(behave_series,axis=1).count()
```

```python
# Grouping by the length of the animal names
animals.groupby(len).sum()
```

```python
keys = ['A','B','A','B']
```

```python
animals.groupby([len,keys]).max()
```
