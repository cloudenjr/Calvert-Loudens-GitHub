```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
# When calling the '.read_csv()' function, MAKE SURE TO PROVIDE ENTIRE PATH or t wont work
csv_file = '/Volumes/LACIE SETUP/Data_Science/Data_Analysis_Viz_InPython/Working_With_Data_Part_3/winequality-red.csv'
dframe_wine = pd.read_csv(csv_file,sep=';')

dframe_wine.head()
```

```python
# Finding the highest alcohol content wine for each quality range
def ranker(df):
    df['alc_content_rank'] = np.arange(len(df)) + 1
    return df
```

```python
dframe_wine.sort_value('alcohol',ascending=False,inplace=True)
```

```python
dframe_wine = dframe_wine.groupby('quality').apply(ranker)
```

```python
num_of_qual = dframe_wine['quality'].value_counts()

num_of_qual
```
