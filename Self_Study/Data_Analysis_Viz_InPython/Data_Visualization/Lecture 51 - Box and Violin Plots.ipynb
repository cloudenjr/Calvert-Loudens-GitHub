# Violin/Viola Plots

- People prefer these plots b/c it combines the simple format of a box plot with the information of a 
KDE plot

```python
# Here the .boxplot() wont give you the entire picture
# Normal Distribution
data1 = stats.norm(0,5).rvs(100)

# Two gamma distributions and concatenate them together
data2 = np.concatenate([stats.gamma(5).rvs(50)-1,
                        -1*stats.gamma(5).rvs(50)])

# Box plot both data1 and data2
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2,bw=0.01)
```

```python
# Idea of bringing a rug plot to the violin plot: each line represents a tick or value
sns.violinplot(data1,inner='stick')
```
