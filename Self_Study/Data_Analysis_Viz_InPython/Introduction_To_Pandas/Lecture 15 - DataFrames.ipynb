```python
import numpy as np  # for array handling

import pandas as pd

from pandas import Series,DataFrame
```

```python
# Python's built-in method to grab a web page
import webbrowser
website='http://en.wikipedia.org/wiki/NFL_win-loss_records'
webbrowser.open(website)
```

```python
# How to copy and read data (from a website) using Pandas

nfl_frame = pd.read_clipboard()
```

```python
nfl_frame
```

```python
# Grabbing column names from your DataFrame in Pandas
nfl_frame.columns
```

```python
# Two ways to grab a column name: 

# 1. Single words
nfl_frame.Team

# 2. Mult Word Columns
nfl_frame['First Season']
```

```python
# Syntax to grab multiple columns; you can define you column parameters after defining the frame you want to 
# work with in the DataFrame() function
DataFrame(nfl_frame,columns=['Team','First','Season','Total','Games'])
```

```python
# Pandas doesnt return an error, when a column passed doesnt exist in the orginal set, just a NULL value
# You have a bunch of NULL values but your DataFrame still works
DataFrame(nfl_frame,columns=['Team','First','Season','Total','Games','Stadium'])
```

```python
nfl_frame
```

```python
nfl_frame.head() # '.head()' function retrieves the first 5 rows, unless specified within the ()

nfl_frame.tail() # '.tail()' function retrieves the last 5 rows, unless specified within the ()
```

```python
# Retrieving rows on a particular index: dataframe_name.loc[]= for label based indexing or 
# '.iloc[]= for positional indexing
nfl_frame.ix[3]
```

```python
# Assigning values to entire columns

nfl_frame['Stadium'] = "Levi's Stadium"
```

```python
nfl_frame
```

```python
nfl_frame['Stadium'] = np.arange(11)

nfl_frame
```

```python
# Adding a series to a DataFrame

stadiums = Series(["Levi's Stadium", "AT&T Stadium"],index=[10,0])

stadiums
```

```python
# How do you input this series into the DataFrame: dataframe_name['column(s) you want'] = your series_name

nfl_frame['Stadium'] = stadiums

nfl_frame
```

```python
# Deleting entire columns: del dataframe_name['column_name]
del nfl_frame['Stadium']
```

```python
nfl_frame
```
