```python
import numpy as np

import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe = DataFrame({'key1':['A']*2 + ['B']*3,
                    'key2':[2,2,2,3,3]})

dframe
```

```python
# How to find duplicates
dframe.duplicated()
```

```python
# How to drop duplicates
dframe.drop_duplicates()
```

```python
# Filtering out duplicates by column
dframe.drop_duplicates(['key1'])
```

```python
dframe
```

```python
# Syntax for taking the last duplicate (bottom up vs. top down)
dframe.drop_duplicates(['key1'],keep='last')
```
