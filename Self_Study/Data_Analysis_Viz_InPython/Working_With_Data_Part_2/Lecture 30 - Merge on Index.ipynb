```python
import numpy as np

import pandas as pd
from pandas import Series,DataFrame
```

```python
df_left = DataFrame({'key':['X','Y','Z','X','Y'],'data':range(5)})

df_left
```

```python
df_right = DataFrame({'group_data':[10,20]},index=['X','Y'])

df_right
```

```python
# Merge using the key for the left dataframe and the index for the right
pd.merge(df_left,df_right,left_on='key',right_index=True)
```

# Using a Hierarchy Index with Merging

```python
df_left_hr = DataFrame({'key1':['SF','SF','SF','LA','LA'],'key2':[10,20,30,20,30],'data_set':np.arange(5.)})
```

```python
df_left_hr
```

```python
df_right_hr = DataFrame(np.arange(10).reshape(5,2),index=[['LA','LA','SF','SF','SF'],[20,10,10,10,20]],
                        columns=['col_1','col_2'])
```

```python
df_right_hr
```

```python
# Merge using the key for the left dataframe and the index for the right but with indexed hierarchies
pd.merge(df_left_hr,df_right_hr,left_on=['key1','key2'],right_index=True)
```

```python
# Another method of merging without using so many arguments: the '.join()' method
df_left.join(df_right)
```
