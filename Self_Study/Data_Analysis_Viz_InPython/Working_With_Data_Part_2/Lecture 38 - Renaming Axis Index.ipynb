```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
```

```python
dframe = DataFrame(np.arange(12).reshape(3,4),index=['NY','LA','SF'],columns=['A','B','C','D'])

dframe
```

```python
# Just like a Series, DataFrame Indexes can utilize the '.map()' function as well
# Here we are making the string of city names (our index) lower case, but not actually assigning 
# it to the dataframe object we created
dframe.index.map(str.lower)
```

```python
# Assigning the new change to the index to the dataframe you created
dframe.index = dframe.index.map(str.lower)
```

```python
dframe
```

```python
# You can also use the '.rename()' function to create a transformed version of your dataframe without modifying
# the original version
dframe.rename(index=str.title,columns=str.lower)
```

```python
# You can also use the '.rename()' function to insert dictionaries providing new values
# The change below is not permanent, to do so, use the arg,'inplace=True' after your columns arg
dframe.rename(index={'ny':'NEW YORK'},columns={'A':'ALPHA'})
```
