```python
import numpy as np

import pandas as pd
from pandas import Series,DataFrame
```

```python
# Creating a dataframe from a dictionary
dframe1 = DataFrame({'key':['X','Z','Y','Z','X','X'],'data_set_1':np.arange(6)})

dframe1
```

```python
dframe2 = DataFrame({'key':['Q','Y','Z'],'data_set_2':[1,2,3]})

dframe2
```

```python
# Many to one merge method; merge will automatically choose overlapping columns to merge on 
pd.merge(dframe1,dframe2)
```

```python
# syntax to match on a specific column
pd.merge(dframe1,dframe2,on='key')
```

```python
# Syntax to choose which dataframes to use and to specify which key(from which dataframe) to use
pd.merge(dframe1,dframe2,on='key',how='left')
```

```python
dframe2
```

```python
dframe1
```

```python
pd.merge(dframe1,dframe2,on='key',how='right')
```

```python
# Syntax to create a union of both keys and use
pd.merge(dframe1,dframe2,on='key',how='outer')
```

```python
# Syntax to merge with multiple keys
df_left = DataFrame({'key1':['SF','SF','LA'],'key2':['one','two','one'],'left_data':[10,20,30]})
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'],'key2':['one','one','one','two'],'right_data':[40,50,60,70]})
```

```python
df_left
```

```python
df_right
```

```python
# Union merge syntax
pd.merge(df_left,df_right,on=['key1','key2'],how='outer')
```

```python
pd.merge(df_left,df_right,on='key1')
```

```python
# Adding suffix to to overcome overlapping(similar) key names when attempting to merge
# In this example, it would be the '_x/_y' defined by key names above
pd.merge(df_left,df_right,on='key1',suffixes=('_lefty','_righty'))
```
